```python
#Uploading dataset
import pandas as pd
import numpy as np
import altair as alt
file_path = "world-data-2023_new.xlsx"
df = pd.read_excel(file_path,sheet_name='world-data-2023_new')
df.head
```

```python
#Identifying missing values
missing_values_summary = df.isnull().sum
print("Missing values summary :\n", missing_values_summary)
# Dropping missing values
threshold = len(df) * 0.2
df_cleaned = df.dropna(axis=1, thresh=threshold)
#imputing missinh values
for column in df_cleaned.columns:
    if df_cleaned[column].dtype in ['int64', 'float64']:
        df_cleaned[column].fillna(df_cleaned[column].median(),inplace=True)
    else:
        df_cleaned[column].fillna(df_cleaned[column].mode()[0],inplace=True)


```
